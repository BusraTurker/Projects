```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup, Comment
#from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
#from selenium import webdriver
import time
import re
```

```python
href_links = []

url="https://www.arabam.com/ikinci-el/otomobil?page="

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' 
                      'AppleWebKit/537.11 (KHTML, like Gecko) '
                      'Chrome/23.0.1271.64 Safari/537.11'}

for i in range(1,5): #The number of data to be retrieved can be adjusted by changing the numbers here. Here are the links on the first four pages.
    url1 = url + str(i)
    req = Request(url = url1, headers = headers)
    html = urlopen(req).read()
    soup = BeautifulSoup(html,"html.parser")
    try:
        href = [div.a.get("href") for div in soup.findAll(class_ = "pr10 fade-out-content-wrapper")]
    except:
        print("None")
    href_links.append(href)
```

```python
href_links = [j for sub in href_links for j in sub]
```

```python
href_links
```

```python
links = []

for i, value in enumerate(href_links):
    url = "https://www.arabam.com/" + value
    links.append(url)
```

```python
links
```

```python
price = []
location = []
features = []
for i, value in enumerate(links):
    link_html = urlopen(value)
    link_soup = BeautifulSoup(link_html, "html.parser")
    #price
    price2 = link_soup.find("span", {"class" : "color-red4 font-semi-big bold fl"})
    if price2 == None:
        price2 = link_soup.find("span", {"class" : "color-red4 font-semi-big bold fl w66"})
        if price2 == None:
            price.append("None")
        else:
            price.append(price2.text) 
    else:
        price.append(price2.text)
    #location
    location2 = link_soup.find(class_ = "one-line-overflow font-default-minus pt4 color-black2018 bold")
    if location2 == None:
        location.append("None")
    else:
        location.append(location2.text)
    #features
    info = link_soup.find("ul",{"class":"w100 cf mt16"})
    if info == None:
        features.append("None")
    else:
        features.append(info.text)
```

```python
df = pd.DataFrame(list(zip(price, location, features)), 
               columns =['Price', 'Location','Info']) 
df
```
