```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup, Comment
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
from selenium import webdriver
import time
```

In this project, according to the results when you type "elbise" (dress) on the search button of a shopping site https://www.trendyol.com; The features of all dresses such as the number of comments, points, number of stars, evaluation numbers were extracted with web scraping methods.  

```python
#Collection of dress links

href_links = []

url="https://www.trendyol.com/sr?q=elbise&qt=elbise&st=elbise&os=1&pi="

headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) ' 
                      'AppleWebKit/537.11 (KHTML, like Gecko) '
                      'Chrome/23.0.1271.64 Safari/537.11'}

for i in range(1,6): #1,20 olacak
    url1 = url + str(i)
    req = Request(url = url1, headers = headers)
    html = urlopen(req).read()
    soup = BeautifulSoup(html,"html.parser")
    try:
        href = [div.a.get("href") for div in soup.findAll(class_ = "p-card-chldrn-cntnr")]
    except:
        print("None")
    href_links.append(href)
```

```python
href_links = [j for sub in href_links for j in sub]
```

```python
href_links
```

```python
products_comment_link = []

for i, value in enumerate(href_links):
    url = "https://www.trendyol.com" + value
    products_comment_link.append(url)
```

```python
#The link to extract the data has been created.
dfsep = pd.DataFrame(list(zip(products_comment_link))) 
locsep = dfsep[0].str.split("?", expand = True)
locsept_list = locsep[1].tolist()
locsep0=locsep[0].tolist()
locsep2=[]
locsepson=[]
for i, text in enumerate(locsept_list):
    locsep22 ="/yorumlar?"+ text
    locsep2.append(locsep22)
    
from functools import partial, reduce
linkss=list(reduce(partial(map, str.__add__), (locsep0, locsep2)))
```

```python
#Sellers data frame were created. The vendors in the link were created by splitting the link using the split method.
seller =[]
df = pd.DataFrame(list(zip(href_links))) 
loc = df[0].str.split("/", expand = True)
seller= loc[1]
seller=seller.astype(str)
seller
```
